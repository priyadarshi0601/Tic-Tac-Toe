```python
window=tk.Tk()
window.title("Tic-Tac-Toe")
window.geometry('250x100')
f1=tk.Frame(window)
# Create a Tkinter variable
tkvar = tk.StringVar(window)

# Dictionary with options
choices = { 'Multiplayer','Singleplayer'}
tkvar.set('Multiplayer') # set the default option

popupMenu = tk.OptionMenu(f1, tkvar, *choices)
tk.Label(f1, text="Choose your opponent").grid(row = 1, column = 1,pady=9)
popupMenu.grid(row = 2, column =1)
button = tk.Button(f1, text="ok", command=select)
button.grid(row=2,column=2)
f1.pack()

window.mainloop()
```
